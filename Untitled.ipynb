```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt, pi, exp
%matplotlib inline
```

```python
x = range(-100,100)
mu = 0
sigma = 20
# crée un lambda qui calcule la fonction de répartition normale (gaussienne)
N = lambda x, mu, sigma : (1/sqrt(2*pi))*exp(-(1/2)*pow((x-mu)/sigma,2))
#y = [N(i,mu,sigma) for i in x]
```

```python
# Affiche une fonction normale centrée pour différentes valeurs de sigma
figure = plt.figure()
plot = plt.plot(x,[N(i,0,10) for i in x])
plot = plt.plot(x,[N(i,0,20) for i in x])
plot = plt.plot(x,[N(i,0,50) for i in x])
xpart = [mu,mu+sigma,mu+2*sigma,mu+3*sigma]
plot = plt.scatter(xpart,[N(i,mu,sigma) for i in xpart])
```
