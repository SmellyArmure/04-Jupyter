```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
### Graphes simples

fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

```python
### Personalisations diverses

# Chanegr la taille de police par dfaut
plt.rcParams.update({'font.size': 15})

fig = plt.figure()
ax = plt.axes()
# Couleur spcifie par son nom, ligne solide
plt.plot(x, np.sin(x - 0), color='blue', linestyle='solid', label='bleu')
# Nom court pour la couleur, ligne avec des traits
plt.plot(x, np.sin(x - 1), color='g', linestyle='dashed', label='vert')
# Valeur de gris entre 0 et 1, des traits et des points
plt.plot(x, np.sin(x - 2), color='0.75', linestyle='dashdot', label='gris')
# Couleur spcifi en RGB, avec des points
plt.plot(x, np.sin(x - 3), color='#FF0000', linestyle='dotted', label='rouge')

# Les limites des axes, essayez aussi les arguments 'tight' et 'equal' 
# pour voir leur effet
plt.axis([-1, 11, -1.5, 1.5]);

# Les labels
plt.title("Un exemple de graphe")

# La lgende est gnre  partir de l'argument label de la fonctio
# plot. L'argument loc spcifie le placement de la lgende
plt.legend(loc='lower left');

# Titres des axes
ax = ax.set(xlabel='x', ylabel='sin(x)')
```

```python
###  Sous-graphiques 

figure = plt.figure()

# graphe pour 10 parties
plot1 = plt.subplot(1, 3, 1)
plt.title("10 parties (courbe)",fontweight="bold")
plt.xlabel("numro de la partie")
plot1.yaxis.set_major_locator(plt.MaxNLocator(1))
plot1.axis([-1, 11, 0, 1])
plot1.yaxis.set_ticklabels(["Perdu","Gagn"])
plot1.plot(np.arange(10),1/(np.arange(10)+1), color = 'blue')

# graphe pour 10000 parties (courbe)
plot2 = plt.subplot(1, 3, 2)
plt.title("10 000 parties (courbe)",fontweight="bold")
plt.xlabel("numro de la partie")
plot2.plot(np.arange(10),np.arange(10)^2, color = 'red')

# graphe pour 10000 parties (nuage de point)
plot3 = plt.subplot(1, 3, 3)
plt.title("10 000 parties (nuage de points)",fontweight="bold")
plt.xlabel("numro de la partie")
plot3.scatter(np.arange(10),np.arange(10)^3, color = 'g')

# gestion des marges
plt.gcf().subplots_adjust(left = 0, bottom = 0.2, right = 3, top = 0.9, wspace = 0.2, hspace = 0.2)
```

```python
### Graphes en barres

plt.title("Comparaison des stratgies",fontweight="bold")
plt.xlabel("numro de la partie")
plt.ylabel("pourcentage de parties gagnes (%)")
plot = plt.bar([1,2,3],[10,50,20], tick_label=["Garder", "Changer", "Alatoire"], color = 'blue')
```

```python
### Personnalisations plus pousses de graphes en barres

x = np.random.randn(1000)

plt.style.use('classic')
fig=plt.figure(figsize=(5,3))
ax = plt.axes(facecolor='#E6E6E6')

# Afficher les ticks en dessous de l'axe
ax.set_axisbelow(True)

# Cadre en blanc
plt.grid(color='w', linestyle='solid')

# Cacher le cadre
# ax.spines contient les lignes qui entourent la zone o les 
# donnes sont affiches.
for spine in ax.spines.values():
    spine.set_visible(False)
    
# Cacher les marqueurs en haut et  droite
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# Nous pouvons personnaliser les tiquettes des marqueurs
# et leur appliquer une rotation
marqueurs = [-3, -2, -1, 0, 1, 2, 3]
xtick_labels = ['A', 'B', 'C', 'D', 'E', 'F']
plt.xticks(marqueurs, xtick_labels, rotation=30)

# Changer les couleur des marqueurs
ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    tick.set_color('gray')
    
# Changer les couleur des barres
ax.hist(x, edgecolor='#E6E6E6', color='#EE6666');
```

```python
### barres d'erreur

x = np.linspace(0, 10, 50)
dy = 2
y = np.sin(x)
ybis = np.sin(x) + dy * np.random.randn(50) # randn : bruit gaussien ; rand : uniforme

plt.rcParams.update({'font.size': 15})

fig = plt.figure()
ax = plt.axes()
plt.plot(x, y, color='blue', linestyle='solid', label='sin non bruit')
plt.plot(x, ybis, color='r', linestyle='dotted', label='sin bruit')
plt.errorbar(x, ybis, yerr=dy, fmt='o', color='black', ecolor='lightgray', elinewidth=3, capsize=0 , label='bruit');

plt.axis('tight'); #'equal'

plt.title("Un exemple de graphe")

plt.legend(loc='lower left');

ax = ax.set(xlabel='x', ylabel='y = sin(x)')

```
