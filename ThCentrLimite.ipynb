```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt, pi, exp
%matplotlib inline
```

```python
x = range(-100,100)
mu = 0
sigma = 20
# crée un lambda qui calcule la fonction de répartition normale (gaussienne)
N = lambda x, mu, sigma : (1/sqrt(2*pi))*exp(-(1/2)*pow((x-mu)/sigma,2))
#y = [N(i,mu,sigma) for i in x]
```

```python
# Affiche une fonction normale centrée pour différentes valeurs de sigma
figure = plt.figure()
plot = plt.plot(x,[N(i,0,10) for i in x])
plot = plt.plot(x,[N(i,0,20) for i in x])
plot = plt.plot(x,[N(i,0,50) for i in x])
xpart = [mu,mu+sigma,mu+2*sigma,mu+3*sigma]
plot = plt.scatter(xpart,[N(i,mu,sigma) for i in xpart])
```

```python
# Affiche les valeurs particulières de la fonction de répartition 
print("La fonction de répartition vaut {} en mu, ...\n\
...et respectivement {}, {} et {} pour mu+sigma, mu+2sigma, mu+3sigma."\
      .format("%4.2f"% (N(mu,mu,sigma)),"%4.2f"% N(mu+sigma,mu,sigma),"%4.3f"% N(mu+2*sigma,mu,sigma),\
              "%4.3f"% N(mu+3*sigma,mu,sigma)))
print("(soient {}, {}, {} et {} % de l'amplitude maximale.)".format(\
        "%4.1f"% (100*N(mu,mu,sigma)/N(mu,mu,sigma)),\
        "%4.1f"% (100*N(mu+sigma,mu,sigma)/N(mu,mu,sigma)),\
        "%4.1f"% (100*N(mu+2*sigma,mu,sigma)/N(mu,mu,sigma)),\
        "%4.1f"% (100*N(mu+3*sigma,mu,sigma)/N(mu,mu,sigma))))
```

```python
###################################################
# Petit exercice perso : somme de 20 dés à 6 faces#
###################################################

def joue(nbe):
    tirages = np.random.randint(6, size=(20,nbe))
    #tirages = np.array([[np.random.randint(1,6) for i in range(0,nbe)]for i in range(0,20)])
    somme = np.sum(tirages, 0) # nouvelle variable aléatoire : la somme des 20 tirages
    return somme

# vérification du tableau de sortie
print(joue(1000).shape)
```

```python
# Affiche les résultats

figure = plt.figure()
plot = plt.hist(joue(1000),100) # histogramme pour 100 catégories
```

```python
# On gagne seulement si on a un total de plus de 100. Quel est le pourcentage d'échecs pour une partie donnée ?

def nbe_echec(nbe,valmin):
     return np.count_nonzero(joue(nbe)<=valmin)

print("Sur 1000 essais, {} sont des échecs.".format(nbe_echec(1000,72)))
```

```python
# Combien de fois, par groupe de 1000 doit-on jouer pour gagner au moins une fois ?

def compte_echecs(nbe,valmin):
    cpt=0
    a=nbe_echec(nbe,valmin)
    while a==nbe:
        a=nbe_echec(nbe,valmin)
        cpt+=1
    return cpt,(nbe-a)

tuple=compte_echecs(1000,80)
print("Cette fois, il a fallu échouer {} parties pour {} gagner fois.".format(tuple[0], tuple[1]))
```

```python
# Combien de fois, en moyenne, sur n parties doit-on jouer de fois pour gagner ?
nbe_parties = 1000
tableau = np.array([compte_echecs(1000,80) for i in range(nbe_parties)])
print("SUR {} ESSAIS :".format(nbe_parties))
print("moyenne du nbe de parties : " + str(np.mean(tableau[:,0])))
print("variance du nbe de parties : " + str(np.var(tableau[:,0])))
print("écart-type du nbe de parties : " + str(np.std(tableau[:,0])))
print("nbe min de parties : " + str(np.min(tableau[:,0])))
print("nbe max de parties : " + str(np.max(tableau[:,0])))
print("nbe de gains moy, min, max : " + str(np.mean(tableau[:,1]))+\
       "," + str(np.min(tableau[:,1]))+ "," + str(np.max(tableau[:,1])))
# pour vérifier à la main...
#print(tableau[:,0])
#print(tableau[:,1])
# Affichage du nbe de fois qu'il faut jouer, 
plot = plt.hist(tableau[:,0],100)
```
